```python
rep = {'Once':1,'Double':2, 'Triple':3,'once':1,'double':2, 'triple':3}
```

```python
replac = {'Dollar':'$', 'dollar':'$', 'One':1, 'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Zero':0,'one':1, 'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'seven':7,'eight':8, 'nine':9, 'zero':0}
```

```python
def repeater(s):
	k = 0
	s = [ss.strip() for ss in s.split()][::-1]

	for i in range(len(s) - 1):
		if s[i - k + 1] in rep:
			s[i - k] *= rep[s[i - k + 1]]

			del s[i - k + 1]
			k += 1

	return ' '.join(s[::-1])
```

```python
def replacer(s):
        s=s.split()
        ll= s[:]
        for i in range(len(s)):
                if s[i] in replac:
                        ll[i]=str(replac[s[i]])

        return repeater( ' '.join(ll)) 
```

```python
def conversion(s):
        return replacer(s)
```
